***ZAD 1.***
Które z następujących zadań wymagają w Twojej opinii inteligencji od człowieka:

**Wypełnianie deklaracji PIT** - wymaga pewnej wiedzy o przepisach podatkowych, ale jest to czynność zautomatyzowana, nie wymaga inteligencji

**Streszczanie tekstu** - wymaga zrozumienia treści, wyciągania kluczowych informacji i syntetyzowania ich, co jest działaniem wymagającym inteligencji językowej; jednak na ten moment zostało to już zautomatyzowane, więc nie wymaga inteligencji

**Tłumaczenie tekstu** - wymaga zrozumienia kontekstu, niuansów językowych i kulturowych, jednak również zautomatyzowane, więc nie wymaga inteligencij

**Klasyfikacja tekstu do kategorii tematycznych** - zautomatyzowane z pomocą algorytmów uczenia maszynowego, które analizują tekst i przypisują go do odpowiednich kategorii, nie wymaga inteligencji

**Odpowiadanie na proste pytania zadawane w języku naturalnym (np. polskim)** - chatboty i asystenci głosowi mogą skutecznie odpowiadać na proste pytania, na ten moment chatGPT prowadzi już zaawansowane rozmowy, nie wymaga inteligencji

**Układanie rozkładu jazdy transportu miejskiego** - nie wymaga inteligencji, może być zautomatyzowane za pomocą systemów optymalizacji i planowania tras, chociaż nadal wymaga ludzkiego nadzoru dla uwzględnienia nieprzewidzianych okoliczności

**Programowanie (pisanie programów komputerowych)** - narzędzia takie jak AI asystenci kodowania mogą wspierać programistów, ale pełna automatyzacja jest bardzo trudna i ograniczona do prostych zadań, częściowo wymaga inteligencji

**„Programowanie” kanałów telewizyjnych** - zautomatyzowane w systemach zarządzania treścią i planowania emisji, nie wymaga inteligencji

**Testowanie oprogramowania** - nie wymaga inteligencji, wiele aspektów testowania oprogramowania może być zautomatyzowane, ale testy eksploracyjne i użyteczności nadal wymagają ludzkiego wkładu

**Komponowanie muzyki** - wymaga inteligencji, kreatywności i wyczucia estetyki, AI może generować muzykę, ale pełna kreatywność i emocjonalne niuanse wciąż są trudne do zautomatyzowania

**Rozwiązywanie układów równań**, **Symboliczne obliczanie pochodnych funkcji**, **Symboliczne całkowanie funkcji** - nie wymaga inteligencji, korzystamy z gotowych wzorów i zasad, zautomatyzowane za pomocą narzędzi matematycznych

**Kierowanie samochodem** - wymaga inteligencji, szybkiego podejmowania decyzji w nagłych, niespodziewanych sytuacjach, ale jest też częściowo zautomatyzowane, ponieważ istnieją już samochody, które same się prowadzą

***ZAD 3.***
Które z poniższych rodzajów komunikacyjnego zachowania człowieka mogą być
obecnie skutecznie imitowane przez sztuczne systemy (odpowiednio oprogramowane
maszyny):

• **rozmowa towarzyska** - częściowo; chatboty i wirtualni asystenci mogą prowadzić proste rozmowy towarzyskie, ale często brakuje im głębszego zrozumienia kontekstu, emocji i niuansów ludzkiej konwersacji

• **dyskusja polityczna** - ograniczone; systemy AI mogą generować odpowiedzi na podstawowe tematy polityczne, ale brakuje im zdolności do głębokiego analizowania, argumentowania i zrozumienia złożonych niuansów politycznych i społecznych

• **dyskusja naukowa** - ograniczone; AI może przetwarzać i generować informacje na tematy naukowe, ale nie jest jeszcze w stanie w pełni uczestniczyć w zaawansowanych, kreatywnych i krytycznych dyskusjach naukowych.

• **odpowiadanie na pytania klientów w telefonicznej infolinii** - tak, do pewnego stopnia; automatyczne systemy odpowiedzi głosowych mogą skutecznie obsługiwać proste pytania, ale bardziej skomplikowane problemy mogą wymagać interwencji człowieka

• **odpowiadanie na pytania klientów w internetowej infolinii** - tak, w dużej mierze; chatboty i systemy AI są szeroko wykorzystywane w obsłudze klienta online, radząc sobie z większością standardowych pytań, jednak złożone lub nietypowe pytania mogą wymagać interwencji człowieka

**Wniosek:**
ChatGPT jest na wyższym poziomie wypowiedzi niż podane przykładowe chatboty.

Odpowiadanie wymijająco, ogólnikowo, częsta zmiana tematu rozmowy, problemy z utrzymaniem wątków - te problemy wystąpiły m.in. w przypadku cleverbota "rozmawiającego" z chatbot.pl.

Zadawanie pytań powstających z moich wypowiedzi, zadawanie dużej liczby pytań - można było zauważyć przy rozmowie z już bardziej inteligentnym chatem GPT.

Nawiązywanie do słów kluczowych - boty asystenty często są lepiej przystosowane do reagowania na słowa kluczowe, ponieważ ich głównym celem jest dostarczanie odpowiednich informacji lub rozwiązań.

Zadawanie pytań - boty „udające człowieka” mogą zadawać więcej pytań, starając się podtrzymać rozmowę. Boty asystenty zadają pytania głównie w celu zebrania informacji potrzebnych do udzielenia odpowiedzi.

**Wnioski:**
Rozmowa z cleverbotem bardzo szybko traci sens, za to chatGPT buduje długie i sensowne wypowiedzi, usuwając jednak kontrowersyjne lub religijne treści.

Zachowanie dwóch rozmawiających ze sobą botów:
Boty mogą popaść w pętlę pytań i odpowiedzi, powtarzając te same frazy, szczególnie jeśli nie mają mechanizmu do rozpoznawania zakończenia wątku.
Mogą szybko zmieniać tematy rozmowy, jeśli oba boty „udające człowieka” nie potrafią utrzymać kontekstu.

Jeśli chodzi o zdenerwowanie bota to poprzez podniesienie poziomu trudności rozmowy (przez zadawanie skomplikowanych pytań lub używanie nietypowych fraz) boty takie jak cleverbot czy chatbot.pl dawały bezsensowną odpowiedź. Próba zdenerwowania bota może prowadzić do powtarzania tych samych fraz lub do wymijających odpowiedzi, co wskazuje na ograniczenia w radzeniu sobie z nieoczekiwanymi sytuacjami. Boty asystenty są mniej podatne na tego rodzaju próby i zwykle dążą do zamknięcia rozmowy w sposób uprzejmy i rzeczowy. ChatGPT przy takich pytaniach albo prosi o więcej informacji albo udziela błędnej odpowiedzi.

**OGÓLNE WNIOSKI:**

Chociaż chatboty osiągnęły znaczny postęp w interakcji z ludźmi, nadal istnieją wyraźne ograniczenia w ich zdolności do prowadzenia głębokich, logicznych i kontekstowych rozmów.
Boty asystenty są bardziej skuteczne w realizacji konkretnych zadań, podczas gdy boty „udające człowieka” mogą być bardziej angażujące w konwersacji, ale często z mniejszą dokładnością w kontekście praktycznym.
ChatGPT na ten moment wg mnie powinien spełniać oczekiwania użytkownika, w szczególności jego wersja płatna, jednak też popełnia drobne błędy. Może być bardzo dobrą pomocą, lecz korzystając z niego powinno się weryfikować jego odpowiedzi.
