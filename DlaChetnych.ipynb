7. Ludzie próbują dokonać zamachu tylko na tych władców, wobec których nie
są lojalni. - VxVy((Władca(y) ^ Człowiek(x)) -> (PróbowałZamachu(x, y) -> -Lojalny(x, y)))
8. Markus próbował dokonać zamachu na Cezara. - PróbowałZamachu(m, c)

**b)** Posługując się powyższymi predykatami spróbuj odpowiedzieć na pytanie, czy
Markus był lojalny wobec Cezara. Czy są one wystarczające/kompletne, by
można było przeprowadzić formalny dowód takiego stwierdzenia? Jeżeli nie,
uzupełnij powyższą listę i podaj dowód.

Z predykatu „Markus próbował dokonać zamachu na Cezara” wynika PróbowałZamachu(m, c). Z predykatu nr 7 „Ludzie próbują dokonać zamachu tylko na tych władców, wobec których nie są lojalni” wynika, że jeśli ktoś próbował dokonać zamachu na władcę, to nie jest lojalny wobec niego. Zatem: PróbowałZamachu(m, c) -> -Lojalny(m, c). Stąd możemy wywnioskować, że Markus nie jest lojalny wobec Cezara. Czy te założenia są wystarczające, aby przeprowadzić formalny dowód? Tak, ponieważ wnioskowanie jest bezpośrednie: próba zamachu oznacza brak lojalności.
